```python
%%time
data = pd.read_csv('E:\BaiduNetdiskDownload/000001.csv')#导入数据并查看数据，查看运行时间
data
```

```python
data['Day'] = pd.to_datetime(data['Day'],format='%Y/%m/%d')#指定日期的输出格式为年/月/日
```

```python
data = data.sort_values(by=['Day'], axis=0, ascending=True)#按找day日期的升序排列/False则是降序排列
```

```python
data.set_index('Day', inplace = True)#建立有关日期的索引
```

```python
data
```

```python
data_new = data['1995-01':'2025-08'].copy()#创建一个原数据1995年1月到2025年8月数据的独立副本，则修改data_new的数据不会对原数据产生影响
data_new
```

```python
data_new['Close'] = pd.to_numeric(data_new['Close'])
data_new['Preclose'] = pd.to_numeric(data_new['Preclose'])#将data_new中的Close列从文本转换为浮点数，preclose列从文本转换为浮点数
```

```python
data_new
```

```python
data_new['Raw_return'] = data_new['Close'] / data_new['Preclose'] - 1#Rt=Pt-Pt-1/Pt-1，收益率的计算公式/日度收益率
```

```python
data_new['Log_return'] = np.log(data_new['Close']) - np.log(data_new['Preclose'])#使用numpy库的对数函数，将收益率转化为对数形式
```

```python
data_new#查看计算结果 
```

```python
data_new1 = data_new.copy()#建立data_new的独立副本
```

```python
data_new1['year'] = data_new1.index.year
data_new1['month'] = data_new1.index.month#建立有关年月的日期索引
```

```python
Month_data1 = data_new1.groupby(['year', 'month'])['Log_return'].sum().to_frame()#按年份和月份分组，计算每组内的对数是收益率的综合
Month_data1['Raw_Return'] = np.exp(Month_data1['Log_return']) - 1#推导原始收益率
```

```python
Month_data1#展示推导结果
```

```python
Quarter_data1 = data_new.resample('QE')['Close'].last().to_frame()#季度收益的计算。/提取每个季度的收盘价并将其转换格式
```

```python
Quarter_data1['Preclose'] = Quarter_data1['Close'].shift(1)#在季度数据中添加preclose列
```

```python
Quarter_data1['Raw_return'] = Quarter_data1['Close'] / Quarter_data1['Preclose'] - 1#计算原始收益率
```

```python
Quarter_data1['Log_return'] = np.log(Quarter_data1['Close']) - np.log(Quarter_data1['Preclose'])#计算对数收益率
```

```python
print("季度末价格计算的收益率:")
Quarter_data1
```

```python
data_new2 = data_new.copy()#建立data_new的数据独立副本
```

```python
data_new2['year'] = data_new2.index.year#建立日期索引
```

```python
Year_data1 = data_new2.groupby('year')['Log_return'].sum().to_frame()#按年份分组计算每个年份内的收益率综合
```

```python
Year_data1['Raw_Return'] = np.exp(Year_data1['Log_return']) - 1#计算原始收益率以及对数收益率
```

```python
print("\n使用groupby计算的年度收益率:")
Year_data1
```

```python
rolling_returns = pd.DataFrame()#初始化dataframe用于储存后续不同窗口的收益率
for window in [5, 10, 20, 30, 60]:#以5，10，20，30，60为滚动窗口的时间窗口
    rolling_log_return = data_new['Log_return'].rolling(window=window).sum()#计算滚动窗口内的原始收益率
    rolling_returns[f'Rolling_{window}d_Return'] = np.exp(rolling_log_return) - 1#计算对应对数收益率
```

```python
print("滚动收益率:")
print(rolling_returns.tail())
```

```python
leiji_returns = pd.DataFrame()#初始化dataframe
leiji_returns['leiji_Log_Return'] = data_new['Log_return'].cumsum()#累加log_return的值计算累计对数率
leiji_returns['leiji_Return'] = np.exp(leiji_returns['leiji_Log_Return']) - 1#计算累计原始收益率
```
